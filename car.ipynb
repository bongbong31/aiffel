- 만약 브랜드명이랑 모델명이 같을 때 Doors랑 Seats 수 같은 값으로 채우기
- 최빈값 구하기
- 매칭하기

```python
# car_df['door_mode'] = car_df.groupby(['brand', 'model'])['Doors'].transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
```

```python
car_df['Doors'] = car_df.groupby(['brand', 'model'])['Doors'].transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
```

```python
# car_df['seat_mode'] = car_df.groupby(['brand', 'model'])['Seats'].transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
```

```python
car_df['Seats'] = car_df.groupby(['brand', 'model'])['Seats'].transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
```

```python
car_df
```

```python
car_df = car_df[~car_df[['Doors', 'Seats']].isnull().any(axis=1)]
```

```python
car_df.info()
```

- Engine NULL 확인 (모델&브랜드 확인 후 fillna)
- Emission NULL 확인 (모델&브랜드&연도 확인 후 fillna)

```python
car_df.groupby(['brand', 'model'])['Engine'].transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
```

```python
car_df['Engine']=car_df.groupby(['brand', 'model'])['Engine'].transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
```

- Engine, Emission Class type object -> float

```python
car_df = car_df[~car_df['Engine'].isnull()]
```

```python
car_df['Emission Class'] = car_df.groupby(['brand', 'model','Registration_Year'])['Emission Class'].transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
```

```python
car_df[car_df['Emission Class'].isnull()].count()
```

```python
car_df = car_df[~car_df['Emission Class'].isnull()]
```

```python
car_df.info()
```

```python
car_df['Emission Class'] = car_df['Emission Class'].str.replace('Euro', '').astype('int')
```

```python
car_df['Engine'] = car_df['Engine'].str.replace('L', '').astype('float')
```
